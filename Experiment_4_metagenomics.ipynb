```python

import os

print("Downloading Experimental Samples...")
# Sample 1
!wget -qnc ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR124/071/SRR12486971/SRR12486971_1.fastq.gz
!wget -qnc ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR124/071/SRR12486971/SRR12486971_2.fastq.gz

# Sample 2
!wget -qnc ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR124/072/SRR12486972/SRR12486972_1.fastq.gz
!wget -qnc ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR124/072/SRR12486972/SRR12486972_2.fastq.gz

print("Unzipping files...")
!gunzip -f *.fastq.gz
print("✅ Data Ready.")
```

```python

samples = ["SRR12486971", "SRR12486972"]

print("Running Metagenomic Analysis...")
for sample in samples:
    print(f"Processing {sample}...")
    !kraken2 --db protocol_db \
             --report {sample}.k2report \
             --report-minimizer-data \
             --minimum-hit-groups 3 \
             {sample}_1.fastq {sample}_2.fastq > {sample}.kraken2

print("✅ Classification Complete. Reports generated.")
```

```python


import os

# List all the report files ending in .k2report
report_files = [f for f in os.listdir('.') if f.endswith('.k2report')]

if len(report_files) == 0:
    print("No reports found to display.")
else:
    for filename in report_files:
        print(f"\n{'='*20}")
        print(f"REPORT FOR: {filename}")
        print(f"{'='*20}\n")

        # Open and print the first 20 lines of the file
        with open(filename, 'r') as f:
            lines = f.readlines()
            # Print the header and top results
            for line in lines[:20]:
                print(line.strip())

        print(f"\n... (Total lines: {len(lines)})")
```
