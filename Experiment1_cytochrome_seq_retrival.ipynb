```python
import os
from google.colab import files
from Bio import SeqIO, Entrez
mysearch = Entrez.esearch(db='nucleotide',term="[cytochrome oxidase]",retmax=5)
myrecord = Entrez.read(mysearch)
mysearch.close()
myids=myrecord["IdList"]
for id1 in myids:
  #Entrez.email='A.N>other@example.com' #always tell NCBI (server for genbank db) who you are
  filename = id1 + ".fasta"
  if not os.path.isfile(filename):                                    # file is created but physical is file presence has to be checked
    net_handle = Entrez.efetch(                                       # Download
        db='nucleotide', id=id1, rettype="fasta", retmode="text"      # Data Retreving
    )
    out_handle= open(filename, "w")                                   # refernce object
    out_handle.write(net_handle.read())
    files.download(filename)
    out_handle.close()
    net_handle.close()
  myrecord = SeqIO.read(filename,"fasta")
  print(myrecord)
```
